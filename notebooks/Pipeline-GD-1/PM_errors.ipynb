```python
import astropy.table as at
import astropy.units as u
import astropy.coordinates as coord
import pyia
import numpy as np
import gala.coordinates as gc
```

```python
data = at.Table.read("/Users/Tavangar/Work/CATS_Workshop/cats/data/joined-GD-1.fits")
```

```python
cat = data[data['phot_g_mean_mag'] <20.5]
cat['pm1'] = cat['pm_phi1_cosphi2_unrefl']
cat['pm2'] = cat['pm_phi2_unrefl']
```

```python
cat
```

```python
tbl = pyia.GaiaData(cat)
```

```python
pm_cov, _ = tbl.get_cov(coords=['pmra', 'pmdec'])
```

```python
c = coord.SkyCoord(ra = cat['ra']*u.deg, dec=cat['dec']*u.deg)
```

```python
frame = gc.GD1Koposov10()
```

```python
gd1_coord_pm_cov = gc.transform_pm_cov(c=c, cov=pm_cov, to_frame=frame)
```

```python
cat['pm1_error'] = np.sqrt(gd1_coord_pm_cov[:,0,0])
cat['pm2_error'] = np.sqrt(gd1_coord_pm_cov[:,1,1])
```

```python
cat.write("/Users/Tavangar/Work/CATS_Workshop/cats/data/joined-GD-1.fits", overwrite=True)
```
